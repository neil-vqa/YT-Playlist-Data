"Youtube playlist: Visualization of sorted videos"

Basic implementation of: 
    1. Youtube API
    2. Pandas Data Frame
    3. Pyplot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from yt_playlist import playlist_serv
show = playlist_serv('https://www.youtube.com/watch?v=iWZmdoY1aTE&list=PLjp0AEEJ0-fGi7bkjrGhBLUF9NMraL9cL', 18)
```

```python
df = pd.DataFrame(show)
df1 = df[['title','views','likes','dislikes']]
df1.head()
```

```python
df1['views'] = df1.views.astype(int)
df1['likes'] = df1.likes.astype(int)
df1['dislikes'] = df1.dislikes.astype(int)
```

```python
vox = df1
v_vox = vox.sort_values(['views'],ascending=False)
v_vox.head()
```

```python
plt.bar(range(v_vox.shape[0]),v_vox['views'], color='#f7dc6f')
plt.xticks(range(v_vox.shape[0]),v_vox['title'],rotation=90)
plt.ylabel('Views')
plt.title('Video Views in the Playlist\n')

plt.show()
```

```python
l_vox = vox.sort_values(['likes'],ascending=False)
l_vox.head()
```

```python
plt.bar(range(l_vox.shape[0]),l_vox['likes'], color='#1E90FF')
plt.xticks(range(l_vox.shape[0]),l_vox['title'],rotation=90)
plt.ylabel('Likes')
plt.title('Video Likes in the Playlist\n')

plt.show()
```

```python
d_vox = vox.sort_values(['dislikes'],ascending=False)
d_vox.head()
```

```python
plt.bar(range(d_vox.shape[0]),d_vox['dislikes'], color='#DE3163')
plt.xticks(range(d_vox.shape[0]),d_vox['title'],rotation=90)
plt.ylabel('Dislikes')
plt.title('Video Dislikes in the Playlist\n')

plt.show()
```
